```python
import numpy as np
import matplotlib.pyplot as plt
```

**Functions kernel**

*   Lineal Kernel

  $Κ(\textbf{x},\textbf{y}) = \textbf{x}^{\top} \textbf{y} $

*   Polynomial kernel

  $Κ(\textbf{x},\textbf{y}) = (\gamma . \textbf{x}^{\top} \textbf{y} + r )^d $

*   Radial Basic Function Kernel

  $Κ(\textbf{x},\textbf{y}) = exp \left( - \frac{|| \textbf{x} - \textbf{y} ||^2}{2 σ^2} \right) $

*   Sigmoid Kernel

  $Κ(\textbf{x},\textbf{y}) = tanh (α . \textbf{x}^{⊤}\textbf{y} + c) $

*   Exponential Kernel

  $Κ(\textbf{x},\textbf{y}) = exp \left( - \frac{|| \textbf{x} - \textbf{y} ||}{2 σ^2} \right) $







```python
def linear_kernel(x, y):
    return np.dot(x, y)

def polynomial_kernel(x, y, degree=3, gamma=1, coef0=0):
    return (gamma * np.dot(x, y) + coef0)**degree

def rbf_kernel(x, y, sigma=0.10):
    return np.exp(-np.linalg.norm(np.array(x) - np.array(y)) / (2 * sigma**2))

def sigmoid_kernel(x, y, alpha=1, coef0=0):
    return np.tanh(alpha * np.dot(x, y) + coef0)

def exponential_kernel(x, y, sigma=0.6):
    return np.exp(-np.linalg.norm(np.array(x) - np.array(y))**2 / (2 * sigma**2))

```

```python
x_range = np.linspace(-5, 5, 500)

fig, ax1 = plt.subplots(figsize=(12, 8))


ax1.plot(x_range, [linear_kernel([1], [x]) for x in x_range], label="Linear", color='b')
ax1.plot(x_range, [polynomial_kernel([1], [x]) for x in x_range], label="Polynomial", color='g')
ax1.plot(x_range, [rbf_kernel([1], [x]) for x in x_range], label="RBF", color='r')
ax1.get_yaxis().set_visible(False)
ax1.get_xaxis().set_visible(False)
ax1.legend(loc="upper left", prop={'size': 22})

ax2 = ax1.twinx()  # Crear un segundo eje y
ax2.plot(x_range, [sigmoid_kernel([1], [x]) for x in x_range], label="Sigmoid", color='m')
ax2.plot(x_range, [exponential_kernel([1], [x]) for x in x_range], label="Exponential", color='y')
ax2.get_yaxis().set_visible(False)

ax2.legend(loc="lower right", prop={'size': 22})

plt.show()
```
