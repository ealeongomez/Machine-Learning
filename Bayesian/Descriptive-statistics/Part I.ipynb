#**Medidas de tendencia central**

$$ Average =  \frac{1}{N} \sum_{i=1}^N x_i  $$ 

$$ Median (odd) = x_{(n+1)/2}^{\text{ordered}} $$

$$ Median (pair) \frac{x_{n/2}^{\text{ordered}} + x_{n/2+1}^{\text{ordered}}}{2} $$

### Mode = $$ x_k $$ where $$ \text{Freq}(x_k) = \max{(\text{Freq}(x_i))} $$ 

```python
df['price_usd'].mean()
```

```python
df['price_usd'].median()
```
