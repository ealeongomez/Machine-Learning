# Joint and marginal provbability

- Joint probability (priduct rule)

$$p(X, Y) = \underbrace{p(X|Y)}_{\text{Conditional}} \; \underbrace{p(B)}_{\text{Marginal}}$$

- Marginal probability

$$p(X) = ∑ p(X,Y) = \sum p(X|Y=y) p(Y=y)$$

```python
X = np.random.normal(size=1000)
Y = np.random.normal(size=1000)

df = pd.DataFrame({'X': X, 'Y': Y})

plt.figure()
sns.jointplot(data=df, x='X', y='Y')
plt.axhline(y=0, color='r', linestyle='-')
plt.axhline(y=1, color='r', linestyle='-')
plt.grid()
plt.show()
```

Conditional probability

$$ p(X|Y)=\frac{p(X,Y)}{p(Y)} $$

$$\text{ if } p(Y)>0$$


```python
value = 1
X_given_y_equals_value = df[(df['Y'] > 0) & (df['Y'] <= 1)]['X']

plt.figure()
sns.displot(x=X_given_y_equals_value)
plt.title("p(X|Y=1)")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Generate random data
data = np.random.normal(0, 1, 1000)

# Calculate the probability density function and the cumulative distribution function
x = np.linspace(-4, 4, 1000)
pdf = norm.pdf(x)
cdf = norm.cdf(x)

# Plot probability density function and cumulative distribution function
plt.figure(figsize=(8, 5))
plt.plot(x, pdf, label='Probability Density Function', color='skyblue')
plt.plot(x, cdf, label='Cumulative Distribution Function', color='salmon')
#plt.title('Probability Distribution Function and Cumulative Function')
plt.xlabel('Value', fontsize=12)
plt.ylabel('Density / Cumulative Probability', fontsize=12)
plt.legend(fontsize=12)
plt.grid(False)
plt.show()

```
