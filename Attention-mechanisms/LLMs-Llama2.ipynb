```python
model_name_or_path = "TheBloke/Llama-2-13B-chat-GGML"
model_basename = "llama-2-13b-chat.ggmlv3.q5_1.bin" # the model is in bin format
```

#**Step 2: Import All the Required Libraries**

```python
from huggingface_hub import hf_hub_download
```

```python
from llama_cpp import Llama
```

#**Step 3: Download the Model**

```python
model_path = hf_hub_download(repo_id=model_name_or_path, filename=model_basename)
```

#**Step 4: Loading the Model**

```python
# GPU
lcpp_llm = None
lcpp_llm = Llama(
    model_path=model_path,
    n_threads=2, # CPU cores
    n_batch=512, # Should be between 1 and n_ctx, consider the amount of VRAM in your GPU.
    n_gpu_layers=32 # Change this value based on your model and your GPU VRAM pool.
    )
```

```python
# See the number of layers in GPU
lcpp_llm.params.n_gpu_layers
```

#**Step 5: Create a Prompt Template**

```python
prompt = """ Separeme dirección y complento de la siguiente dirección:

CALLE 45 # 86 -38 UNIDAD RINCON DEL SUR APTO 302 TORRE B

Toma de ejemplo los siguientes casos:

- BOSQUES DE PUENTE PALMA CALLEA 3RA 63A 46 SECTOR D APTO 307 TORRE 3
Direccion: CALLEA 3RA 63A 46
Complemento: BOSQUES DE PUENTE PALMA SECTOR D APTO 307 TORRE 3

- CALLE 50 # 59-72 HOGAR GERIATRICO ALBA
Direccion: CALLE 50 # 59-72
Complemento: HOGAR GERIATRICO ALBA
"""

prompt_template=f"""SISTEMA: Eres un asistente servicial, respetuoso y honesto. Responde siempre de la forma más amable.

USER: {prompt}

ASISTENTE:
"""

print(prompt_template)
```

#**Step 6: Generating the Response**

```python
response=lcpp_llm(prompt=prompt_template, max_tokens=256, temperature=0.5, top_p=0.95,
                  repeat_penalty=1.2, top_k=150,
                  echo=True)
```
